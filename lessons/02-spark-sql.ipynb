```python
from pyspark import SparkConf, SparkContext
from pyspark.sql import SparkSession
```

```python
conf = SparkConf() \
      .setMaster("local[3]") \
      .setAppName("InitialRDD")

spark = SparkSession \
      .builder \
      .config(conf=conf) \
      .getOrCreate()
```

```python
btc_df = spark.read \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .csv("./data/coin_Bitcoin.csv")
```

```python
btc_df.show()
```

```python
btc_df.createOrReplaceTempView("btc_trades_tbl")
count_data = spark.sql("select count(Symbol) from btc_trades_tbl")
count_data.show()
```

```python
spark.stop()
```
